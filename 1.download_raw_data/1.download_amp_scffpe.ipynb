```python
# Load required libraries
import synapseclient
import synapseutils
from pathlib import Path
import csv
```

```python
syn = synapseclient.Synapse()
syn.login()
```

```python
base = Path('/data/srlab/AMP_collab/data/early_disease_synovium/scFFPE')
data = [
    ['syn68205076', base / 'scFFPE_RA-SYN-EDP1/Level_2_scFFPE'],
    ['syn68205074', base / 'scFFPE_PsD-SYN-EDP1/Level_2_scFFPE']]
```

```python
# download recursively
for syn_id, target in data:
    print(syn_id, '->', target)
    try:
        Path(target).mkdir(parents=True, exist_ok=True)
        all_files = synapseutils.syncFromSynapse(syn, entity=syn_id, path=target)
    except FileExistsError as e:
        print(e)
        pass
```
