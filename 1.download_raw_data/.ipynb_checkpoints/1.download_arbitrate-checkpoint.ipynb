```python
# Load required libraries
from zenodo_get import download 
from pathlib import Path
import csv
import shutil, re
```

```python
output_dir = Path('/data/srlab/AMP_collab/data/early_disease_synovium/xenium/Xenium_RA-SYN-ARBITRATE/Level_1_Xenium_Output/')
data = [
    ['10.5281/zenodo.15058261', output_dir], 
    ['10.5281/zenodo.15058714', output_dir]
]
```

```python
# download recursively
for syn_id, target in data:
    print(syn_id, '->', target)
    try:
        Path(target).mkdir(parents=True, exist_ok=True)
        download(syn_id, output_dir = target)
    except FileExistsError as e:
        print(e)
        pass
```

```python
base = Path(output_dir)
pattern = re.compile(r"(patient\d+)")

for d in base.iterdir():
    if not d.is_dir():
        continue
    match = pattern.search(d.name)
    if not match:
        continue  # skip dirs without a patient ID
    patient_id = match.group(1)
    dst = output_dir / patient_id 
    dst.parent.mkdir(parents=True, exist_ok=True)
    shutil.move(str(d), str(dst))
```
