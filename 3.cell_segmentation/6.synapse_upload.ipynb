```python
import os
from datetime import date
import synapseclient
import synapseutils
from synapseclient import Project

syn = synapseclient.Synapse()
syn.login()
```

# Upload control files

```python
# Edit this variable to set the destination synID in Synapse
remote_project_id = 'syn74490454'

# Edit this variable to set the path to the local data
LOCAL_PROJECT_DIRECTORY = '/data/srlab/AMP_collab/data/early_disease_synovium/xenium/Xenium_CTRL-SYN-EDP1/Level_3'
print(LOCAL_PROJECT_DIRECTORY)

# This variable sets a name and location for the file manifest, recommend choosing a separate directory from the data
LOCAL_FILE_MANIFEST = './control_manifest.tsv'
print(LOCAL_FILE_MANIFEST)
```

```python
# Step 2: Create a manifest TSV file to upload data in bulk
# Note: When this command is run it will re-create your directory structure within
# Synapse. Be aware of this before running this command.
# If folders with the exact names already exists in Synapse, those folders will be used.
synapseutils.generate_sync_manifest(
    syn=syn,
    directory_path=LOCAL_PROJECT_DIRECTORY,
    parent_id=remote_project_id,
    manifest_path=LOCAL_FILE_MANIFEST
)
```

```python
# Step 3: After generating the manifest file, we can upload the data in bulk
synapseutils.syncToSynapse(
    syn=syn, manifestFile=LOCAL_FILE_MANIFEST, sendMessages=True
)
```
